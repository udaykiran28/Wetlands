```julia
using CSV, DataFrames, DelimitedFiles
data_read = CSV.read("/Users/udaykiran/Library/Mobile Documents/com~apple~CloudDocs/Wetlands/Marshes/Dataset/Hoskote_Lake/Hoskote_birds.csv", DataFrame)

```

```julia
data_copy = copy(data_read)
```

```julia
#rename!(data_copy, Symbol("COMMON NAME") => :"COMMON_NAME")
rename!(data_copy, Symbol("OBSERVATION DATE") => :"OBSERVATION_DATE_YEAR_ONLY")
```

```julia
#show(println(names(data_copy)))
```

```julia
using Dates

data_copy[!, "OBSERVATION_DATE_YEAR_ONLY"] = year.(data_copy[!, "OBSERVATION_DATE_YEAR_ONLY"])
data_copy
```

```julia
using Chain, Missings


result_CPochard = @chain data_copy begin
    subset(_, "COMMON_NAME" => ByRow(x -> !ismissing(x) && x == "Common Pochard"))
    groupby(_, "OBSERVATION_DATE_YEAR_ONLY")
    combine(_, nrow => :count)
end
```

```julia
# plot the trend for Common Pochard, Cotton-Pygmy Goose, Eurasian Wigeon, Garganey, Green-Winged Teal and Indian Spot Billed Duck. 

using PlotlyJS

CPochard_plot() = scatter(
    x=result_CPochard.OBSERVATION_DATE_YEAR_ONLY,
    y=result_CPochard.count,
    mode="lines+markers",
    marker=attr(
        size=9,
        color="blue",
        symbol="circle",
        line=attr(width=2, color="lightorange")
    ),
    line=attr(color="lightbrown", width=2)
)

CPGoose_plot() = scatter(
    x=result_CPGoose.OBSERVATION_DATE_YEAR_ONLY,
    y=result_CPGoose.count,
    mode="lines+markers",
    marker=attr(
        size=9,
        color="black",
        symbol="lightbrown",
        line=attr(width=2, color="lightbrown")
    ),
    line=attr(color="brown", width=2)
)

EWigeon_plot() = scatter(
    x=result_EWigeon.OBSERVATION_DATE_YEAR_ONLY,
    y=result_EWigeon.count,
    mode="lines+markers",
    marker=attr(
        size=9,
        color="black",
        symbol="circle",
        line=attr(width=2, color="grey")
    ),
    line=attr(color="yellow", width=2)
)

garganey_plot() = scatter(
    x=result_Garganey.OBSERVATION_DATE_YEAR_ONLY,
    y=result_Garganey.count,
    mode="lines+markers",
    marker=attr(
        size=9,
        color="black",
        symbol="circle",
        line=attr(width=2, color="grey")
    ),
    line=attr(color="brown", width=2)
)
GWTeal_plot() = scatter(
    x=result_GWTeal.OBSERVATION_DATE_YEAR_ONLY,
    y=result_GWTeal.count,
    mode="lines+markers",
    marker=attr(
        size=9,
        color="black",
        symbol="circle",
        line=attr(width=2, color="grey")
    ),
    line=attr(color="pink", width=2)
)
ISBDuck_plot() = scatter(
    x=result_ISBDuck.OBSERVATION_DATE_YEAR_ONLY,
    y=result_ISBDuck.count,
    mode="lines+markers",
    marker=attr(
        size=9,
        color="black",
        symbol="circle",
        line=attr(width=2, color="lightbrown")
    ),
    line=attr(color="green", width=2)
)
plot_CPochard = plot(CPochard_plot(), Layout(title="Common Pochard"))
plot_CPGoose = plot(CPGoose_plot(), Layout(title="Cotton Pygmy-Goose"))
plot_EWigeon = plot(EWigeon_plot(), Layout(title="Eurasian Wigeon"))
plot_garganey = plot(garganey_plot(), Layout(title="Garganey"))
plot_GWTeal = plot(GWTeal_plot(), Layout(title="Green-Winged Teal"))
plot_ISBDuck = plot(ISBDuck_plot(), Layout(title="Indian Spot-billed Duck"))

plot_piscivorous1 = [plot_CPochard plot_CPGoose plot_EWigeon; plot_garganey plot_GWTeal plot_ISBDuck]

relayout!(plot_piscivorous1, 
    height = 500,
    width = 700,
    title="Population trend of Migrants of Wetland birds around Hoskote Lake",
    showlegend = false,   
    )

plot_piscivorous1
```
