```julia
# Read the CSV file 

using CSV, DataFrames, DelimitedFiles
read_data = CSV.read("/Users/udaykiran/Library/Mobile Documents/com~apple~CloudDocs/Wetlands/Marshes/Dataset/Hoskote_Lake/Hoskote_birds.csv", DataFrame)
```

```julia
# Make a copy of the file

data_copy = copy(read_data)
```

```julia
println(sum(ismissing.(data_copy.FAMILY)))
```

```julia
# Data from the source (.csv) is already filtered baed on the keyword, "Hoskote". 
# Rename columns by adding underscore '_' between the names to avoid issues related to space. 
rename!(data_copy, Symbol("OBSERVATION DATE") => :"OBSERVATION_DATE_YEAR_ONLY")

```

```julia
println(sum(ismissing.(data_copy.OBSERVATION_DATE_YEAR_ONLY)))
```

```julia
using Dates

# filter the column: OBSERVATION_DATE_YEAR_ONLY with only year and store in the same column. 
data_copy[!, "OBSERVATION_DATE_YEAR_ONLY"] = year.(data_copy[!, "OBSERVATION_DATE_YEAR_ONLY"])
data_copy
```

```julia
# Remove missing values in FAMILY
# Accessing data of Acrocephalidae family

using Chain, Missings


result_Acrocephalidae = @chain data_copy begin
    subset(_, "FAMILY" => ByRow(x -> !ismissing(x) && x == "Acrocephalidae"))
    groupby(_, "OBSERVATION_DATE_YEAR_ONLY")
    combine(_, nrow => :count)
end
```
