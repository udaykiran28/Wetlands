```julia
using CSV, DataFrames, DelimitedFiles
data_read = CSV.read("/Users/udaykiran/Library/Mobile Documents/com~apple~CloudDocs/Wetlands/Marshes/Dataset/Hoskote_Lake/Hoskote_birds.csv", DataFrame)

```

```julia
data_copy = copy(data_read)
```

```julia
rename!(data_copy, Symbol("COMMON NAME") => :"COMMON_NAME")
#rename!(data_copy, Symbol("OBSERVATION DATE") => :"OBSERVATION_DATE_YEAR_ONLY")
```

```julia
using Dates

data_copy[!, "OBSERVATION_DATE_YEAR_ONLY"] = year.(data_copy[!, "OBSERVATION_DATE_YEAR_ONLY"])
data_copy
```

```julia
# Remove missing values in COMMON_NAME
# Accessing data of Ashy Woodswallow 

using Chain, Missings


result_AWSwallow = @chain data_copy begin
    subset(_, "COMMON_NAME" => ByRow(x -> !ismissing(x) && x == "Ashy Woodswallow"))
    groupby(_, "OBSERVATION_DATE_YEAR_ONLY")
    combine(_, nrow => :count)
end
```
